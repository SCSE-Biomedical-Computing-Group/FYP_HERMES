```python
import pandas as pd
```

```python
younger_volume_df = pd.read_csv('combined.csv')
younger_volume_df = younger_volume_df.drop(younger_volume_df.columns[0], axis=1)
younger_volume_df
```

```python
elder_volume_df = pd.read_csv('elderCombined.csv')
elder_volume_df = elder_volume_df.drop(elder_volume_df.columns[1], axis=1)
elder_volume_df.columns.values[0] = 'sub_id'
elder_volume_df
```

```python
combined_df = pd.concat([younger_volume_df, elder_volume_df], ignore_index=True)
combined_df
```

```python
file_path = 'patientsData.csv'
combined_df.to_csv(file_path, index=False)
```
